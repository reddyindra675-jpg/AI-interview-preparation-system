```python
# AI Interview Preparation System

score = 0

questions = {
    "What is Python?": "programming language",
    "What is a variable?": "store data",
    "What is a loop?": "repeat"
}

for question, keyword in questions.items():
    answer = input(question + "\nYour Answer: ").lower()

    if keyword in answer:
        print("Correct!\n")
        score += 1
    else:
        print("Needs Improvement.\n")

print("Final Score:", score, "/", len(questions))

if score == len(questions):
    print("Excellent Performance!")
elif score >= 2:
    print("Good Job!")
else:
    print("Keep Practicing!")
```
